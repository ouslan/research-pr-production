```python
import bambi as bmb
from prprod_repl import ProdRepl
import polars as pl
import arviz_plots as azp
import arviz as az

azp.style.use("arviz-variat")


pr = ProdRepl()
```

```python
df = pr.reg_data()
df = df.with_columns(
    capital_synth_log=pl.col("capital_synth").log(),
    labour_log=pl.col("labour").log(),
    energy_log=pl.col("energy"),
).sort(["year", "month", "muni"])
df = df.to_pandas()
df
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# 1. Load the data
# Assuming your dataframe is named 'df'
df['date'] = pd.to_datetime(df[['year', 'month']].assign(day=1))

# 2. Aggregate energy by date
# This sums energy across all municipalities for each month
energy_trend = df.groupby('date')['energy_log'].sum().reset_index()

# 3. Create the visualization
plt.figure(figsize=(10, 5))
plt.plot(energy_trend['date'], energy_trend['energy_log'], marker='s', color='#2ca02c', linewidth=2)

# Formatting
plt.title('Total Energy Consumption by Date', fontsize=14)
plt.xlabel('Year-Month', fontsize=12)
plt.ylabel('Aggregate Energy Units', fontsize=12)
plt.grid(True, linestyle='--', alpha=0.7)
plt.tight_layout()

plt.show()
```
